###Train-Test Split

```python
x = df[['time', 'Frequency(Kmph)', 'Weight(Kg)']]
y = df['Temperature']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

###Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Initialize and fit the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the test set
y_pred = model.predict(x_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Temperature")
plt.ylabel("Predicted Temperature")
plt.title("Linear Regression: Actual vs Predicted Temperature")
plt.show()
```

```python
diff=y_test-y_pred
```

```python
import seaborn as sns
sns.distplot(diff)
```

###Decision-Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize the Decision Tree Regressor and fit the model
decision_tree = DecisionTreeRegressor()
decision_tree.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred_dt = decision_tree.predict(x_test)

# Calculate metrics for Decision Tree
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print(f"Decision Tree - Mean Squared Error: {mse_dt}")
print(f"Decision Tree - R-squared: {r2_dt}")
```

```python
# Visualizing Actual vs Predicted for Decision Tree
plt.scatter(y_test, y_pred_dt)
plt.xlabel("Actual Temperature")
plt.ylabel("Predicted Temperature (Decision Tree)")
plt.title("Decision Tree: Actual vs Predicted Temperature")
plt.show()
```

###Random-Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor and fit the model
random_forest = RandomForestRegressor()
random_forest.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred_rf = random_forest.predict(x_test)
```

```python
# Calculate metrics for Random Forest
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Random Forest - Mean Squared Error: {mse_rf}")
print(f"Random Forest - R-squared: {r2_rf}")
```

```python
# Visualizing Actual vs Predicted for Random Forest
plt.scatter(y_test, y_pred_rf)
plt.xlabel("Actual Temperature")
plt.ylabel("Predicted Temperature (Random Forest)")
plt.title("Random Forest: Actual vs Predicted Temperature")
plt.show()
```

```python
# Additional analysis for the Random Forest model, such as feature importance:
feature_importance = random_forest.feature_importances_

# Display or plot the feature importance
for i, importance in enumerate(feature_importance):
    print(f"Feature {i + 1} Importance: {importance}")
```

```python
# Define a list of feature names
features = ['Feature 1 Name', 'Feature 2 Name', 'Feature 3 Name']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(features, feature_importance)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance in Random Forest Model')
plt.xticks(rotation=45)
plt.show()
```

###Training, Evaluation, and Feature Importance for XGBoost

```python
from xgboost import XGBRegressor

# Initialize the XGBoost Regressor and fit the model
xgboost = XGBRegressor()
xgboost.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred_xgboost = xgboost.predict(x_test)
```

```python
# Calculate metrics for XGBoost
mse_xgboost = mean_squared_error(y_test, y_pred_xgboost)
r2_xgboost = r2_score(y_test, y_pred_xgboost)

print(f"XGBoost - Mean Squared Error: {mse_xgboost}")
print(f"XGBoost - R-squared: {r2_xgboost}")
```

```python
# Additional analysis for the XGBoost model, such as feature importance:
feature_importance_xgboost = xgboost.feature_importances_

# Display or plot the feature importance
for i, importance in enumerate(feature_importance_xgboost):
    print(f"Feature {i + 1} Importance: {importance}")

# Define a list of feature names
features = ['Feature 1 Name', 'Feature 2 Name', 'Feature 3 Name']

plt.figure(figsize=(10, 6))
plt.bar(features, feature_importance_xgboost)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance in XGBoost Model')
plt.xticks(rotation=45)
plt.show()
```
